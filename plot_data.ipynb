```python
import pandas as pd
import plotly.express as px

from utils.data_handler import DataHandler
```

```python
get_data = DataHandler("src/Pokemon.csv")
```

```python
collection = get_data.get_whole_collection()
```

Pick up columns to use in the plot.

```python
columns = ["identifier", "name", "type", "total_stats", "generation", "legendary"]

dataframe = pd.DataFrame.from_dict(collection)
dataframe = dataframe[columns]

if not dataframe.empty:
    print("Dataframe loaded.")
```

Configure and show the plot.

```python
map_colors = {
    "Grass": "#0f9d58",
    "Fire": "#e53238",
    "Water": "#4285f4",
    "Bug": "#A4C636",
    "Normal": "#AF6E4D",
    "Poison": "#75378d",
    "Electric": "#d4c53a",
    "Ground": "#d3a674",
    "Fairy": "#ef64bc",
    "Fighting": "#872626",
    "Psychic": "#63436d",
    "Rock": "#95441b",
    "Ghost": "#8329a9",
    "Ice": "#68c3ee",
    "Dragon": "#137e4e",
    "Dark": "#353539",
    "Steel": "#818181",
    "Flying": "#adf9ec"
}

figure = px.scatter(
    title="Pokemon Type/Stats",
    data_frame=dataframe,
    x="identifier",
    y="total_stats",
    color="type",
    hover_name="name",
    hover_data=["type", "total_stats", "generation", "legendary"],
    color_discrete_map=map_colors,
)

figure.show()
```
